## Embeddings are generated by "Multimodal_unsupervised_pretraining"
Download unsupervised trained embedding (PBMC160k_multimodal_concerto_embedding.h5ad) for PBMC160K(Protein + RNA) from 10.6084/m9.figshare.19243578

```python
import scanpy as sc
import pandas as pd
from sklearn.metrics.cluster import normalized_mutual_info_score, adjusted_rand_score, silhouette_score
```

## embedding generated by Concerto

```python
conc_2 = sc.read_h5ad('PBMC160k_multimodal_concerto_embedding.h5ad')
```

```python
conc_2
```

## Clustering by Leiden

```python
sc.pp.neighbors(conc_2, use_rep='X')
```

```python
res = 0.08
sc.tl.leiden(conc_2, resolution = res, key_added = f"leiden_{res}")
```

```python
nmi_ = normalized_mutual_info_score(conc_2.obs['label'], conc_2.obs[f"leiden_{res}"])
ari_ = adjusted_rand_score(conc_2.obs['label'], conc_2.obs[f"leiden_{res}"])
```

```python
print(nmi_, ari_)
```
